## 1. Berapa banyaknya yang meminjam sepeda ketika weekday?
## 2. Berapa banyaknya sepeda yang dipinjam selama 5 jam ketika weekday?

### import library

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Menginput data csv

```python
df = pd.read_csv("hour.csv",delimiter=",")
```

### Menghitung jumlah sepeda yang dipinjam ketika weekday, workingday, dan holiday

```python
#Banyaknya yang meminjam ketika weekday
weekday = df['weekday'].sum()
weekday
```

```python
workingday = df['workingday'].sum()
```

```python
holiday = df['holiday'].sum()
```

```python
day =('Holiday', 'Weekday', 'Workingday')
peminjam = (holiday, weekday, workingday)
```

```python
plt.bar(x=day, height=peminjam)
plt.title('Peminjam Sepeda ketika Holiday, Weekday, dan Workingday\n')
plt.ylabel('Banyak Sepeda')
plt.show()
```

## Banyaknya yang meminjam sepeda ketika weekday sebanyak 52201 sepeda
